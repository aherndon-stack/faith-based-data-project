## 📥 Step 1: Load the Playlist CSV

````python
import pandas as pd

# Load the dataset
df = pd.read_csv("praise_playlist.csv")

# Show the first few rows
df.head()
````

## 🧼 Step 2: Clean + Understand the Data
Let’s get some basic info about what’s inside.

````python
# Overview of the data
df.info()

# Check for missing values
df.isnull().sum()
````

## 🔢 Step 3: Count the Songs by Mood

````python
# Count how many songs there are for each mood
df['Mood'].value_counts()
````

## 📊 Step 4: Visualize It!
Let’s make a chart showing which moods are most common.

````python
import matplotlib.pyplot as plt

# Bar chart of moods
df['Mood'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Songs by Mood')
plt.xlabel('Mood')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
````

## 🧠 Step 5: Average Song Length
Let’s find out how long these worship songs usually are.

````python
# Average length
df['Length_Minutes'].mean()
# Summary statistics for song length
df['Length_Minutes'].describe()
````
